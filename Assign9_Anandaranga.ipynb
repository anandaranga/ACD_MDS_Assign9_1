```python
# Count positional distance to the previous zero, or the start of the Series, whichever is closer

import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})

# make a new column with zeros at zeros and nans elsewhere
df = df.assign(Dist =df.loc[df.X==0])

# use cumsum() to for each group that start after a zero
nul_value= df['Dist'].isnull()
df.assign(Dist = nul_value.groupby((nul_value.diff() == 1).cumsum()).cumsum())
```

```python
#Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers

import numpy as np
import pandas as pd

# create a series of 261 random numbers for each business day in 2015; set the seed to help validate results
np.random.seed(100)
df = pd.DataFrame({'s':np.random.randint(1, 1000, 261)})
df.head()
```

```python
# add a column that contains DateTimeIndex of each business day of 2015
start_date = '1/1/2015'
end_date = '12/31/2015'
df = df.assign(DateTimeStamp = pd.bdate_range(start_date, end_date))
df.shape[0]
```

```python
# assign this column to be the index for the DataFrame
df.set_index ('DateTimeStamp', inplace=True)
df
```

```python
#Find the sum of the values in s for every Wednesday (Note - Monday = 0 and Sunday = 6, so Wednesday = 2)
df['Weekday'] = df.index.weekday
df.groupby('Weekday').sum()
```

```python
sum_of_wed = df.groupby('Weekday').get_group(2).sum()
sum_of_wed
```

```python
# Average For each calendar month
df['CalendarMonth'] = df.index.month
df.groupby('CalendarMonth').s.mean()
```

```python
# Highest value for each calendar month
df.groupby('CalendarMonth').s.max()
```

```python
# Find the date on which the highest value occurred for each month
df[df['s'] == df.groupby(['CalendarMonth'])['s'].transform(max)]
```
